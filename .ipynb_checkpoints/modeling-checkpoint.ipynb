```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns

```

```python
model_df = pd.read_csv('model.csv', index_col='AssignedID')
data = pd.read_csv('data.csv', index_col='AssignedID')
```

```python
model_df
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve, auc

```

```python
model_df.Purpose.value_counts()
```

### Classification
The target column here has values 0f (1) for loans that are for purchase and (2) for refinance.  For the sake of this classification problem, I will be classifying if the loan is for purchase or not.  Since the column only has two unique data values, if the loan is not for purchase, then it must be for refinance, therefore (1) will represent loans for purchase, and (0) will represent refinancing loans.

```python
# change vlaues to represent (1) for purchasing and (0) for refinancing

model_df.Purpose.replace(to_replace=2, value=0, inplace=True)
model_df.Purpose.value_counts()
```

```python
# fit a classifier model 
X = model_df.drop(columns=['Purpose'])
y = model_df.Purpose

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= .30, random_state=31)

clf = xgb.XGBClassifier()
clf.fit(X_train, y_train)
training_preds = clf.predict(X_train)
val_preds = clf.predict(X_test)
training_accuracy = accuracy_score(y_train, training_preds)
val_accuracy = accuracy_score(y_test, val_preds)

print("Training Accuracy: {:.4}%".format(training_accuracy * 100))
print("Validation accuracy: {:.4}%".format(val_accuracy * 100))
```

```python
# Get AUC/ROC for train/test
y_train_score = clf.predict_proba(X_train)[:,1]

y_test_score = clf.predict_proba(X_test)[:,1]


fpr_train, tpr_train, thresh_train = roc_curve(y_train, y_train_score)

fpr, tpr, thresh = roc_curve(y_test, y_test_score)
```

```python
# plot ROC curve

sns.set_style('darkgrid', {'axes.facecolor': '0.9'})

print('Test-AUC: {}'.format(auc(fpr, tpr)))
print('Train-AUC: {}'.format(auc(fpr_train, tpr_train)))
plt.figure(figsize=(10, 8))
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='Test-ROC curve')
plt.plot(fpr_train, tpr_train,
         lw=lw, label='Train-ROC curve')
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.yticks([i/20.0 for i in range(21)])
plt.xticks([i/20.0 for i in range(21)])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

###  Model Results

Based on the AUC and ROC, the model predicted accurately whether or not the loan was for purchasing or refinancing.  The model generalized well when taking in new data as well.  With an accurate model in place, I can now dive into the feature importances to determine which features were useful.

```python
# get feature importance scores and put in dataframe
feat_imp = pd.DataFrame(clf.feature_importances_, columns=['Importance'], index=X.columns)

feat_imp.Importance.describe()

```

```python
# subset of top half of importances
importance = feat_imp[feat_imp.Importance >= feat_imp.Importance.median()]

# plot importances

importance.plot.barh(figsize=(13,11))
plt.title("Feature Importance")
plt.xlabel('Percentage')
plt.ylabel('Features')

plt.show()
```

### Feature Importance:

Initially, looking at what stands out the most is the "First" feature, which denotes if the borrower is a first time home buyer or not.  It makes sense that being a first time home buyer would be important in predicting if a borrower is purchasing or refinancing since a first time home buyer will only be purchasing and not refinancing.

Another observation about the feature importances is that besides the "First" feature, there aren't very many features that had much usefulness.  There's only a total of 6 features with an importance greater than 0.05.  These will make up the top features in the next subset.

```python
# subset only top features with importance percentage > 0.05

top_imp = feat_imp[feat_imp.Importance >= 0.05]

top_imp.plot.barh(figsize=(13,11))
plt.title("Feature Importance")
plt.xlabel('Percentage')
plt.ylabel('Features')

plt.show()
```
